```python
data = raw
```

Inster values into DF

```python
df = pd.DataFrame(data = data.split('\n'))
df.columns = ['row']
df_matrix = df['row'].str.split('', expand=True)
df_matrix.loc[len(df)] = '.'
top_row = df_matrix[len(df):]
df_matrix = pd.concat([top_row, df_matrix]).reset_index(drop = True)
df_matrix
```

Set up variables/function to search matrix

```python
rows = np.arange(0, len(df_matrix), 1, dtype = int)
columns = np.arange(0, df_matrix.shape[1], 1, dtype = int)
```

```python
def is_integer(n):
    try:
        float(n)
    except ValueError:
        return False
    else:
        return float(n).is_integer()
```

```python
#create df to hold location of number with a special character near by
special_number_locations = pd.DataFrame(columns = ['x', 'y'])
x_offset = np.arange(-1,2,1)
y_offset = np.arange(-1,2,1)
```

# Part 1



search for numbers next to special character

```python

row = 0

#loop to search for number with special character next to it
for y in columns:
  for x in rows:
  #check current cell
    if df_matrix.loc[x,y] == '.':
      continue
    elif df_matrix.loc[x,y] == '':
      continue
    elif is_integer(df_matrix.loc[x,y]):
    #if number check surrounding cells
      for xx in x_offset:
        for yy in y_offset:
          if df_matrix.loc[(x-xx),(y-yy)] == '.':
            continue
          elif is_integer(df_matrix.loc[(x-xx),(y-yy)]):
            continue
          elif df_matrix.loc[(x-xx),(y-yy)] == '':
            continue
          else:
            special_number_locations.loc[row] = [(x), (y)]
            row += 1

```

```python
special_number_locations
```

```python
special_number_start_end = pd.DataFrame(columns = ['x_start', 'y_start', 'x_end', 'y_end'])
```

Find start and end of numbers

```python
number_rows = np.arange(0, len(special_number_locations), 1, dtype = int)

for row in number_rows:
  number_x = special_number_locations.loc[row,'x']
  number_y = special_number_locations.loc[row,'y']
  i = 0
  e = 0
  #find first number
  while (df_matrix.loc[number_x,number_y-i] != '.') and ((number_y-i) != 0) and (df_matrix.loc[number_x,(number_y-i)] != '') and is_integer(df_matrix.loc[number_x,number_y-i]):
    i = i+1

    #if (df_matrix.loc[((number_x-i)-1),number_y] == '.') or ((number_x-i-1) == 0) or (df_matrix.loc[((number_x-i)-1),number_y] == ''):
    #  special_number_start_end[row, 'x_start'] = number_x
    #  special_number_start_end[row, 'y_start'] = (number_y-i)
    #  i =-1
    #else:
    #  i =-1
  #find last number
  while (df_matrix.loc[number_x,number_y+e] != '.') and ((number_y+e) != df_matrix.shape[1]) and (df_matrix.loc[number_x,number_y+e] != '') and is_integer(df_matrix.loc[number_x,number_y+e]):
    #if (df_matrix.loc[number_x,(number_y+e+1)] == '.') or ((number_y+e+1) == df_matrix.shape[1]) or (df_matrix.loc[number_x,(number_y+i+e)] == ''):
    #  special_number_start_end[row, 'x_end'] = number_x
    #  special_number_start_end[row, 'y_end'] = (number_y+e)
    #  e =+1
    #else:
    e = e+1
  special_number_start_end.loc[row] =[number_x,number_y-i+1,number_x, number_y+e-1]
```

Drop repeats

```python
special_number_start_end=special_number_start_end.drop_duplicates(subset= ['x_start', 'y_start', 'x_end', 'y_end'])
special_number_start_end=special_number_start_end.reset_index()
```

```python
special_number_rows = np.arange(0, len(special_number_start_end), 1, dtype = int)

for row in special_number_rows:
  x_row = special_number_start_end.loc[row,'x_start']
  y_start2 =special_number_start_end.loc[row,'y_start']
  y_end2 =special_number_start_end.loc[row,'y_end']
  special_number_start_end.loc[row,['number']] = ''.join(df_matrix.iloc[x_row,y_start2:(y_end2+1)].astype(str))
```

Sum numbers

```python
special_number_start_end['number'] = pd.to_numeric(special_number_start_end['number'])
special_number_start_end['number'].sum()
```
