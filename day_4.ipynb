```python
df_winning_num = df_matrix_clean['winning_num']
df_winning_num.columns = ['winning_num']
df_winning_matrix = df_winning_num.str.split(' ', expand=True)
df_winning_matrix
```

```python
df_your_num = df_matrix_clean['your_nums']
df_your_num.columns = ['your_nums']
df_your_num
```

```python
rows = np.arange(0, len(df_winning_matrix), 1, dtype = int)
columns = np.arange(0, df_winning_matrix.shape[1], 1, dtype = int)
```

```python
x= 2
y= 1
row_wins = 0
winning_number = df_winning_matrix.loc[x,y]
wins = df_your_num.loc[x:].str.count(fr'\s{winning_number}(?![\d])').iloc[0]
row_wins = row_wins + wins
doubled_score = 2^(row_wins)
row_wins
#results.loc[x] = [(x+1), (row_wins),(doubled_score)]
```

```python
results = pd.DataFrame(columns = ['gameID', 'row_wins', 'doubled_score'])
for x in rows:
  row_wins = 0
  doubled_score = 0
  for y in columns:
    if df_winning_matrix.loc[x,y] == '':
      continue
    else:
      winning_number = df_winning_matrix.loc[x,y]
      wins = df_your_num.loc[x:].str.count(fr'\s{winning_number}(?![\d])').iloc[0]
      row_wins = row_wins + wins
      if row_wins == 0:
        doubled_score == 0
      else:
        doubled_score = 2**(row_wins-1)
      results.loc[x] = [(x+1), (row_wins),(doubled_score)]
```
